```python
import numpy as np
import time
```

```python
A = np.random.randint(0, 10, (1_000_000, 8))
print(A.shape, A.dtype)
```

```python
# add all the numbers in the first column
```

```python
start = time.time()
result = A[:, 0].sum()
end = time.time()
print(result)
print((end-start)*1000, "ms")
```

```python
B = A.T.copy().T # B has consecutive columns
```

```python
start = time.time()
result = B[:, 0].sum()
end = time.time()
print(result)
print((end-start)*1000, "ms")
```
